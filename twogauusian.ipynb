```julia
using Plots
using LaTeXStrings
```

```julia
function gauss(θ , σ , μ)
    return (1/sqrt(2*π*σ^2))*exp(-(θ-μ)^2/(2σ^2))
end
```

```julia
function twogauss(θ , σ , μ1 , μ2)
    return 2*gauss(θ , σ , μ1) + gauss(θ , σ , μ2)
end
```

```julia
θ = -5:0.01:5
```

```julia
σ = 1
μ1 = -2
μ2 = 2
y(θ) = twogauss(θ , σ , μ1 , μ2)
```

```julia
plot(
    θ,
    y,
    xlabel = L"\theta",
    ylabel = L"y",
    label = "σ = $σ  , μ1 = $μ1 , μ2 = $μ2",
    title = "Two Gaussian"
)
```

```julia
σ = 1
μ1 = -4
μ2 = 4
y(θ) = twogauss(θ , σ , μ1 , μ2)
```

```julia
plot!(
    θ,
    y,
    xlabel = L"\theta",
    ylabel = L"y",
    label = "σ = $σ  , μ1 = $μ1 , μ2 = $μ2",
    title = "Two Gaussian"
)
```

```julia
θ = -5:0.01:5
```

```julia
p = plot()
for σ in 1:2
    for μ1 in 0:1
        for μ2 in 0:1
            y(θ) = twogauss(θ , σ , μ1 , μ2)
            plot!(θ,y,xlabel=L"\theta",ylabel=L"y",label="σ=$σ,μ1=$μ1,μ2=$μ2",legend=:outertopright)
        end
    end
end
p
```
