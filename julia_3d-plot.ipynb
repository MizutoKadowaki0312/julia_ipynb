```julia
using Plots
```

```julia
f(x,y) = exp(-(x^2+y^2))
```

```julia
x = -2:0.1:2
y = -2:0.1:2
```

```julia
z = f.(x,y)
```

```julia
p1 = heatmap(x,y,f,xlabel="x",ylabel="y",title="heatmap")
p2 = contour(x,y,f,xlabel="x",ylabel="y",title="contour") #ã‚³ãƒ³ã‚¿ãƒ¼å›³ = ç­‰é«˜ç·š
p3 = wireframe(x,y,f,xlabel="x",ylabel="y",zlabel="f(x,y)",title="wireframe")
p4 = surface(x,y,f,xlabel="x",ylabel="y",zlabel="f(x,y)",title="surface")

plot(p1,p2,p3,p4,layout=(2,2),size=(800,600))
```

```julia
x = -2:0.1:2
y = -2:0.1:2
f(x,y) = exp(-(x^2+y^2))
z = f.(x,y)
plot3d(x,y,z)
```

```julia
p = plot3d()
for a in 1:10
    f(x,y) = sin(a*x) + cos(y)
    z = f.(x,y)
    plot3d!(x,y,z,label="a=$a")
end

p
```
