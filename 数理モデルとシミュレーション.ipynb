## 自由落下の運動方程式

$$
v(t) = -gt + v_0
$$
$$
x(t) = v_0t - \dfrac{1}{2}gt^2
$$

```julia
t = 0:0.1:10
```

```julia
v1(t) = v0 - gconst*t
```

```julia
h1(t) = v0*t - (0.50)*gconst*t^2
```

```julia
gconst = 9.8
```

```julia
p1 = plot()

for v0 in 1:10
    v1(t) = v0 - gconst*t
    plot!(t , v1 , label = "v0 = $v0")
end

xlabel!(L"t")
ylabel!(L"v(t)")
title!(L"v(t) = v_0 - gt")

p2 = plot()

for v0 in 1:10
    h1(t) = v0*t - (0.50)*gconst*t^2
    plot!(t , h1 , label = "v0 = $v0")
end

xlabel!(L"t")
ylabel!(L"x(t)")
title!(L"x(t) = x_0 + v_0t - \dfrac{1}{2}gt^2")

plot(p1 , p2 , layout = (2,1) , legend=:outertopright)
```

## 調和振動

x(t) = x_0\cos(\omega t) + \dfrac{v_0}{\omega}\sin(\omega t)
$$

```julia
h2(t) = x0*cos(ω*t) + (v0/ω)*sin(ω*t)
```

```julia
t = 0:0.1:2π
```

# 微分方程式のシミュレーション

##  自由落下モデル

連立微分方程式

$$
\dfrac{dx}{dt} = v(t)
$$

$$
\dfrac{dv}{dt} = -g
$$

を差分方程式を用いて解く．初期条件は，$v(0) = 1.0 , x(0) = 1.0$.

微分方程式を厳密に解くと，

$$
v(t) = -gt + v_0
$$
$$
x(t) = x_0 + v_0t - \dfrac{1}{2}gt^2
$$


差分法についての証明はGoodNotes．

```julia
n = 100
Δt = 0.01
```

```julia
#解を代入する配列を作成．
v2 = zeros(n)

#初速を代入
v2[1] = 1.0
```

```julia
x2 = zeros(n)
x2[1] = 1.0
```

```julia
for i ∈ 1:n-1
    v2[i+1] = v2[i] - gconst*Δt
end
```

```julia
for i ∈ 1:n-1
    x2[i+1] = x2[i] + v2[i]*Δt
end
```

```julia
#=
    時間軸の設定．
    t = 0 から 100要素分なので，99まで．
=#

t = 0:Δt:(n-1)Δt
```

```julia
#差分法による結果のプロット(速度)
p1 = plot(t , v2 , label = L"v_{n+1} = v_n - g\Delta{t}")

#厳密解のプロットとの重ね合わせ
v3(t) = v2[1] - gconst*t
plot!(t , v3 , label =L"v(t) = v_0 - gt")

xlabel!(L"t")
ylabel!(L"v(t)")
title!("Δt = $Δt")

#差分法による結果のプロット(速度)
p2 = plot(t , x2 , label = L"x_{n+1} = x_n + v_n\Delta{t}")

#厳密解のプロットとの重ね合わせ
x3(t) = x2[1] + v2[1]*t - (0.50)*gconst*t^2
plot!(t , x3 , label =L"x(t) = x_0 + v_0t - \dfrac{1}{2}gt^2")

xlabel!(L"t")
ylabel!(L"x(t)")
title!("Δt = $Δt")

plot(p1 , p2 , layout = (2,1) , legend=:outertopright)
```

## Euler 法

### 例題①

$$
\dfrac{dx(t)}{dt} = x
$$
$$
x(0) = 1
$$

を解く．Euler法についての証明はGoodNotes．

厳密に解くと，

$$
x(t) = \exp(t)
$$

となる．

```julia
Δt = 0.01
n = 100
t = 0:Δt:(n-1)*Δt
```

```julia
ex1 = zeros(n)
ex1[1] = 1.0
```

```julia
for i ∈ 1:n-1
    ex1[i+1] = (1 + Δt)*ex1[i]
end
```

```julia
plot(t , ex1 , label = "Euler method , Δt = $Δt")
plot!(t , exp.(t) , label = "Exact Solution" , legend =:best)
xlabel!(L"t")
ylabel!(L"x(t)")
```

#### $Δt$ による違い

### 例題②

```julia
n = 100
Δt = 0.01
t = 0:Δt:(n-1)*Δt
ex5 = zeros(n)
ex5[1] = 1.0
```

```julia
n = 100
Δt = 0.01
t = 0:Δt:(n-1)*Δt

ex5 = zeros(n)
ex5[1] = 1.0

a = 1.0

for i in 1:n-1
    ex5[i+1] = (1-a*Δt)*ex5[i]
end
p1 = plot(t , ex5 , label = "(a,Δt) = ($a,$Δt)")
plot!(t , exp.(-a*t) , label = "exact solution")
```

```julia
n = 100
Δt = 0.01
t = 0:Δt:(n-1)*Δt
ex6 = zeros(n)
ex6[1] = 1.0
p = plot()

@manipulate for a in 0:300
    for i in 1:n-1
        ex6[i+1] = (1-a*Δt)*ex6[i]
    end
    plot!(t , ex6 , label = "a = $a" , legend=:outertopright)
    title!("y = ex6[1]*(1 - $a*Δt)^(x-1) , Δt = $Δt")
end
```

```julia
p = plot()
anim = @animate for a ∈ 0:5:300
    for i in 1:n-1
        ex6[i+1] = (1-a*Δt)*ex6[i]
    end
    plot!(t , ex6 , label = "a = $a" , legend=:outertopright)
    title!("y = ex6[1]*(1 - $a*Δt)^(x-1) , Δt = $Δt")
end
@time gif(anim, "geometric-progression.gif", fps = 80)
```
