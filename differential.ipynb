# 考え方

ここでは，Julialang を用いて差分法を実装する．差分法は微分1階の微分方程式を近似的に解く手段で，関数$f(x)$について

$$
f^{\prime}(x) = \dfrac{f(x + \Delta x) - f(x)}{\Delta x}
$$

となることを用いてコードを組み立てていく．ここで $\Delta x$ は刻み幅だが，微分の定義を考えると $\Delta x$ は 0 に近い方が精度は良くなる．

## ここで扱う例

ここでは，$f(x) = \sin(x)$ を例にとって考えていきます．$\sin(x)$ は微分すると

$$
f^{\prime}(x) = \cos(x)
$$

となります．以下で実装していきましょう．

# 実装

## package の呼び出し

```julia
using Plots
```

## $x$ の範囲

以下，扱っていく関数の定義域を設定します．ここでは，$-\pi \le x \le \pi$ にして，なんとなく0.01 刻みで設定します．

```julia
x = -π:0.01:π
```

## 扱う関数の定義

```julia
f(x) = sin(x)
```

## 微分の刻み幅を設定

```julia
Δx = 0.0001
```

## 差分法による式を定義

```julia
f′(x) = (f(x+Δx) - f(x))/Δx
```

## 可視化

```julia
plot(x , f , lw = 2 , title = "graph", label = "sin(x)")
```

```julia
plot!(x , f′ , lw = 2 , title = "f(x) and the derivative of f(x) in x", label = " the derivative of f(x) in x")
```

## 厳密解との比較

```julia
g(x) = cos(x)
```

```julia
plot!(x , g , ls = :dash , lw = 3 , title = "f(x) and the derivative of f(x) in x and cos(x)" , label = "cos(x)")
```

これにて，差分法が検証された．
