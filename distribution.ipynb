```julia
using Plots
using LaTeXStrings
```

```julia
f(x,λ) = λ^x * exp(-λ) / factorial(BigInt(x))
x = 0:1:40
y1 = f.(x , 20)
plot(x , y1 , marker = "o" , label=L"0\leq x \leq 40")
title!(L"f(x , \lambda) = \frac{\lambda^{x}}{x!}\exp(-\lambda) , \lambda = 20")

x = 0.0:1:40.0
y2 = f.(x , 20)
plot!(x , y2 , marker = "o" , label=L"0.0\leq x \leq 40.0")
```
