# パラメータを連続変化させた関数のプロット

```julia
x = 0:0.1:10
```

```julia
#=
p = plot()
f(x) = x

for a ∈ 1:10
    if a == 1
        f(x) = a*x
        p = plot(x , f , label = "f(x) = $a x" , legend =:topleft)
    else
        f(x) = a*x
        plot!(p , x , f , label = "f(x) = $a x" , legend =:topleft)
    end
end

xlabel!("x")
ylabel!("f(x)")

p
=#
```

```julia
p = plot()

for a ∈ 1:10
    f(x) = a*x^2
    p = plot!(x , f , label = "f(x) = $a x^2" , legend =:top)
end

xlabel!("x")
ylabel!("f(x)")
p
```

# 任意パラメータにおける関数のプロット

```julia
x = -2:0.1:2
param = [-2 , -1 , 1 , 2]
p = plot()

for a ∈ param
    f(x) = a*x^2
    p = plot!(x , f , label = "f(x) = $a x^2")
end

xlabel!("x")
ylabel!("f(x)")

p
```
