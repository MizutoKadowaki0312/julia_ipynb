# Julia で Euler法

## 考え方

ここでは，Euler法を Julia で実行していく方法についてみていく．Euler法は，関数$f(x)$の微分が

$$
f^{\prime}(x) = \dfrac{f(x+\Delta x) - f(x)}{\Delta x}
$$

であることを用いる．この式を変形すると

$$
f(x + \Delta x) = f(x) + f^{\prime}(x)\Delta x
$$

となる．これをいわゆる配列を用いて実装していく．

# 実装

## package の呼び出し

```julia
using Plots
```

## 計算回数の設定

後に使用するfor文で何回計算を回すかを，ここで設定する．

```julia
N = 10000
```

## 関数の定義

ここでは例題として，

$$
y^{\prime} = 3y + 2
$$

をEuler法で見ていこう．

### 初期化

```julia
y = zeros(N)
```

### 初期値

微分方程式を解くので，初期値を設定する．

```julia
y[1] = 1
```

### 計算を回す

#### 刻み幅の設定

```julia
dx = 0.0001
```

#### $x$の範囲を設定

```julia
x = 0:dx:(N-1)dx
```

### for文の組み立て

```julia
for n = 1:N-1
    y[n+1] = (1 + 3*dx)y[n] + 2*dx
end
```

# 厳密解の設定

数値解と厳密解を fit させて，精度を見たい．用意した微分方程式を初期条件 $y(0) = 1$ の下で解くと，

$$
y(x) = \dfrac{5}{3}e^{3x} - \dfrac{2}{3}
$$

となる．

```julia
f(x) = 5/3 * exp.(3x) - 2/3
```

# 可視化

```julia
plot(x , y , lw = 5 , ls = :dash , label = "Numerical" , xlabel = "x" , ylabel = "y")
```

```julia
plot!(x , f , lw = 2 ,  label = "Exact" , xlabel = "x" , ylabel = "y")
```
